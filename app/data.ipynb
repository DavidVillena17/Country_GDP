```python
import pandas as pd
```

# Extracción de información

```python
df = pd.read_csv('data/data_gdp.csv')
df.head() 
```

```python
#Viztazo rápido de la información
df.info()
```

# Limpieza de datos

```python
#Eliminamos Columnas innecesarias
df_cleaned = df.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'])

#Convertimos la columna a tipo de dato String
df_cleaned = df_cleaned.astype({'Country Name': 'string'})

#Realizamos transpose
df_cleaned = df_cleaned.set_index('Country Name')
df_cleaned = df_cleaned.transpose()
df_cleaned.index.name = 'Year'
df_cleaned.head()
```

```python
#Vistazo rápido de existencia de valores nulos
df_cleaned.isna().sum()
```

```python
#Revisamos la existencia de columnas (Paises) sin datos
empty_columns = df_cleaned.columns[df_cleaned.isna().all()].tolist()
print(f'Número de países sin datos: {len(empty_columns)}')
print('Países sin ningún dato:')
print(empty_columns)
```

```python
#Eliminamos columnas sin valores
df_cleaned = df_cleaned.dropna(axis=1, how='all')
```

```python
#Graficamos los países con al menos 50% de sus datos faltantes

import matplotlib.pyplot as plt

porcentaje = df_cleaned.isna().mean() * 100
porcentaje[porcentaje > 50].sort_values(ascending=False).plot(kind='bar', figsize=(16,5))
plt.title('% de datos faltantes por país')
```

```python
#Eliminamos columnas en el dataFrame
columns_to_drop = porcentaje[porcentaje > 50].index
df_cleaned = df_cleaned.drop(columns=columns_to_drop)
```

```python
#Verificamos
porcentaje = df_cleaned.isna().mean() * 100
porcentaje[porcentaje > 50].index.tolist()
```
