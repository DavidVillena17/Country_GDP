```python
import pandas as pd
```

```python
df = pd.read_csv('data/data_gdp.csv')
df.head() 
```

```python
df.info()
```

```python
df = df.astype({'Country Name': 'string', 'Country Code': 'string','Indicator Name':'string', 'Indicator Code':'string'})
df.info()
```

```python
df_transposed = df
df_transposed = df_transposed.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'])
#df_transposed.rename(columns={'Country Name': 'Year'}, inplace=True)
df_transposed = df_transposed.set_index('Country Name')
df_transposed = df_transposed.transpose()
df_transposed.index.name = 'Year'
df_transposed.head()
```
